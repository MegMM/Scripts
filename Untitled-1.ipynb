```python
from pathlib import Path, WindowsPath

class Library:
    def __init__(self, path):
        self.path = WindowsPath(path)
        self.name = self.path.stem
        self.author_dirs = [a for a in self.path.iterdir() if a.is_dir() else ""]

path = Path(r"C:\Users\megha\OneDrive\Calibre Libraries\calibre_laptop_03-25-2019")
library = Library(path)
print(f"{library.path}")
print(f"{library.name}")
print(f"{library.author_dirs}")

```
